```python
locations = df[['lat', 'long']]

map = folium.Map(location=[df.lat[0], df.long[0]], zoom_start=14, control_scale=True)
```

```python
import datetime as dt

df['date'] = pd.to_datetime(df['date'])

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

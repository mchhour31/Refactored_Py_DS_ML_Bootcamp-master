### Adding EarlyStopping to help combat overfitting

It's quite hard to know exactly what the most optimal number of training passes we should do, without validation performance being sacrificed.

One possibly soln is to train the model multiple times with different values, and then select the number of epochs that result in the best performance on the train or holdout set. However, this is computationally inefficient and time consuming..

A better soln is to use **early stopping** -- train the model once for some arbitrarily large number of epochs, and once the performance on the validation set begins to degrade, the training process will stop and the resulting model is the one we can begin to utilise

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid')) # binary classification (if multiclass = softmax)

model.compile(loss='binary_crossentropy',
              optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=True, patience=25)
```

```python
model.fit(X_train, y_train, epochs=600, validation_data=(X_test, y_test),callbacks=[early_stop])
```

```python
pd.DataFrame(model.history.history).plot()
```

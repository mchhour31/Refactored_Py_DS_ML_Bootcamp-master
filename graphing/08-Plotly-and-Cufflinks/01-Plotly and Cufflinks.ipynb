# Plotly and Cufflinks

Plotly is a library that allows you to create interactive plots that you can use in dashboards or websites (you can save them as html files or static images).

## Installation

In order for this all to work, you'll need to install plotly and cufflinks to call plots directly off of a pandas dataframe. These libraries are not currently available through **conda** but are available through **pip**. Install the libraries at your command line/terminal using:

    pip install plotly
    pip install cufflinks

** NOTE: Make sure you only have one installation of Python on your computer when you do this, otherwise the installation may not work. **

## Imports and Set-up

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import chart_studio.plotly as py

print(__version__) # requires version >= 1.9.0
```

```python
import cufflinks as cf
```

```python
# For Notebooks
init_notebook_mode(connected=True) # connects pandas, plotly to an interactive js library
```

```python
# For offline use
cf.go_offline() # using cufflinks for offline use
```

```python
# data
df = pd.DataFrame(np.random.randn(100, 4), columns='A B C D'.split())
```

```python
df
```

```python
df2 = pd.DataFrame({"Category": ['A', 'B', 'C'], 'Values': [32, 43, 50]})
```

```python
df2
```

```python
df.iplot()
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers', color='red', size=10)
```

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df.iplot(kind='box', mode='markers')
```

```python
cf.datagen.lines().iplot(kind='scatter')
```

```python
cf.datagen.lines(3).iplot(kind='scatter')
```

```python
cf.datagen.lines(3).iplot()
```

```python
cf.datagen.lines(2).iplot(kind='scatter', bestfit=True, colors=['blue'], bestfit_colors=['pink', 'red'])
```

```python
cf.datagen.lines(2).iplot(kind='scatter', mode='markers', colorscale='paired')
```

```python
cf.datagen.lines(5).resample('M').mean().iplot(kind='bar')
```

```python
cf.datagen.lines(5).resample('M').mean().iplot(kind='bar', barmode='stack')
```

```python
cf.datagen.box().iplot(kind='box')
```

```python
cf.datagen.box(6).iplot(kind='box')
```

```python
cf.datagen.histogram(2).iplot(kind='histogram', opacity=.75)
```

```python
cf.datagen.heatmap(30, 30).iplot(kind='heatmap')
```

```python
cf.datagen.bubble(prefix='industry').iplot(kind='bubble',x='x',y='y',size='size',categories='categories')
```

```python
cf.datagen.scatter3d(2, 150, mode='stocks').iplot(kind='scatter3d', x='x', 
                                                  y='y', z='z', size=5, categories='categories',
                                                  colors=['blue','pink'], margin=(0,0,0,0))
```

```python
cf.datagen.bubble3d(5, 4, mode='stocks').iplot(kind='bubble3d',x='x', y='y', z='z', size='size', margin=(0,0,0,0))
```

```python
cf.datagen.sinwave(10, .25).iplot(kind='surface')
```
